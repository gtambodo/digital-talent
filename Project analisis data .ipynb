```python
# 1. Deskripsi Data 

import pandas as pd
print ("Terlampir data eksport Indonesia ke berbagai belahan dunia dari tahun 2005-2013 ")
nilai_eksport= pd.read_csv('processed-nilai-ekspor-negara-tujuan-2005-2013.csv', delimiter = ',', encoding = "ISO-8859-1")

nilai_eksport # Deskripsi Data 
# nilai_eksport.columns
```

```python
nilai_eksport.count()
```

```python
# 2. Statistik Data 

# Rata-rata nilai eksport pertahun dari tahun  2005 - 2013 

print ("Rata-rata nilai eksport pertahun dari tahun  2005 - 2013 ")
nilai_eksport.groupby('tahun').mean()

```

```python
# 2. Statistik Data 

# nilai eksport berdasarkan Region

nilai_eksport.groupby('region').mean()
```

```python
# 3. visualisasi nilai eksport berdasarkan tahun 

ax = nilai_eksport.groupby('tahun').mean()
ax = ax.plot(kind='bar',title ="rata-rata berdasarkan tahun", legend=True,fontsize=12)
ax.set_xlabel("Tahun",fontsize=12)
ax.set_ylabel("Nilai Eksport",fontsize=12)
```

```python
# 3. visualisasi eksport berdasarkan region 


ax = nilai_eksport.groupby('region').mean()
ax = ax.plot(kind='bar',title ="rata-rata berdasarkan Region", legend=True,fontsize=12)
ax.set_xlabel("Region",fontsize=12)
ax.set_ylabel("Jumlah Ekport",fontsize=12)
```
